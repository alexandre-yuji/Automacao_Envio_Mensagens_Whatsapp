Enviar mensagem

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from time import sleep
import pyautogui
from openpyxl import load_workbook # trabalhar com excel


nome_arquivo_contatos = "Contatos.xlsx"
planilhaDadosContato = load_workbook(nome_arquivo_contatos)
sheet_selecionada = planilhaDadosContato['Dados'] # usar a aba com nome Dados


navegadorChrome = webdriver.Chrome()
navegadorChrome.get("https://web.whatsapp.com/")

#Aguarda até encontrar o ID side do whatsapp
#Enquanto o tamanho da lista for menor do que 1 fica tentando logar
while len(navegadorChrome.find_elements(By.ID, 'side')) < 1:
    #Aguarda 3 segundos para ver se o Whatsapp Web logou
    sleep(3)
    
#Aguarda 3 segundos para dar tempo do site processas as informações 
sleep(3)

for linha in range(2, len(sheet_selecionada['A']) + 1): # começo na segunda linha da coluna A, até a última linha
    
    #Criamos as variaveis nome e mensagem
    nomeContato = sheet_selecionada['A%s' % linha].value # percorro linha por linha da coluna A armazenando o valor
    mensagemContato = sheet_selecionada['B%s' % linha].value # percorro linha por linha da coluna B armazenando o valor
    
    #Busca pelo XPATH e escreve o nomeContato no campo pesquisa do WhatsApp Web
    navegadorChrome.find_element(By.XPATH, '//*[@id="side"]/div[1]/div/div/div[2]/div/div[1]/p').send_keys(nomeContato)
    sleep(3)
    pyautogui.press('enter')
    sleep(3)

    #Busca pelo XPATH o campo onde escreve a mensagem
    navegadorChrome.find_element(By.XPATH, '//*[@id="main"]/footer/div[1]/div/span[2]/div/div[2]/div[1]/div[2]/div[1]/p').send_keys(mensagemContato)
    sleep(3)
    pyautogui.press('enter')
    sleep(3)
```

Enviar imagem com mensagem

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from time import sleep
import pyautogui
from openpyxl import load_workbook # trabalhar com excel


nome_arquivo_contatos = "Contatos.xlsx"
planilhaDadosContato = load_workbook(nome_arquivo_contatos)
sheet_selecionada = planilhaDadosContato['Dados'] # usar a aba com nome Dados


navegadorChrome = webdriver.Chrome()
navegadorChrome.get("https://web.whatsapp.com/")

#Aguarda até encontrar o ID side do whatsapp
#Enquanto o tamanho da lista for menor do que 1 fica tentando logar
while len(navegadorChrome.find_elements(By.ID, 'side')) < 1:
    sleep(3)
sleep(3)

imagem = "teste.png"

for linha in range(2, len(sheet_selecionada['A']) + 1): # começo na segunda linha da coluna A, até a última linha
    
    #Criamos as variaveis nome e mensagem
    nomeContato = sheet_selecionada['A%s' % linha].value # percorro linha por linha da coluna A armazenando o valor
    mensagemContato = sheet_selecionada['B%s' % linha].value # percorro linha por linha da coluna B armazenando o valor
    
    #Busca pelo XPATH e escreve o nomeContato no campo pesquisa do WhatsApp Web
    navegadorChrome.find_element(By.XPATH, '//*[@id="side"]/div[1]/div/div/div[2]/div/div[1]/p').send_keys(nomeContato)
    sleep(3)
    pyautogui.press('enter')
    sleep(3)

    #Clicar no botão de anexos
    navegadorChrome.find_elements(By.CSS_SELECTOR, "span[data-icon='attach-menu-plus']")
    sleep(2)
    navegadorChrome.find_element(By.CSS_SELECTOR, "span[data-icon='attach-menu-plus']").click()
    
    #Clico no anexo de imagens
    anexo = navegadorChrome.find_element(By.CSS_SELECTOR, "input[type='file']")
    sleep(1)
    anexo.send_keys(imagem)
    sleep(3)
    
    #Busca pelo XPATH o campo onde escreve a mensagem da imagem
    navegadorChrome.find_element(By.XPATH, '//*[@id="app"]/div/div/div[3]/div[2]/span/div/span/div/div/div[2]/div/div[1]/div[3]/div/div/div[2]/div[1]/div[1]/p').send_keys(mensagemContato)
    sleep(3)
    pyautogui.press('enter')
    sleep(3)
```
